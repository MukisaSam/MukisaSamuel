```python
import numpy as np
import cv2
from sklearn.cluster import KMeans
```

```python
# ----------------------------------------------------------
# Image Names
# (Used only for displaying results)
# ----------------------------------------------------------
image_names = [
"dog_1.png","dog_2.png","dog_3.png",
"dog_4.png","dog_5.png","dog_6.png",
"cat_1.png","cat_2.png","cat_3.png",
"cat_4.png","cat_5.png","cat_6.png",
"car_1.png","car_2.png","car_3.png",
"car_4.png","car_5.png","car_6.png",
"mountain_1.png","mountain_2.png","mountain_3.png",
"mountain_4.png","mountain_5.png","mountain_6.png",
"unkn1own_.png",
"unkn2own_.png",
"unkn3own_.png",
"unkn4own_.png",
"unkn5own_.png",
"unkn6own_.png"
]
```

```python
#Obtaining the features from the images
#color and texture features using OpenCV and scikit-image libraries
def extract_features(image_path):
    # 1. Load image
    img = cv2.imread(image_path)
    if img is None:
        return None
        
    # 2. Extract Color: Convert to HSV and get the average Hue value
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    mean_hue = cv2.mean(hsv)[0] # First element is Hue (0-179)
    
    # 3. Extract Texture: Use Laplacian variance to measure edge density
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    laplacian = cv2.Laplacian(gray, cv2.CV_64F)
    texture_variance = laplacian.var() # Measures sharpness/roughness
    
    # Return both numbers as a single pair
    return [round(mean_hue, 2), round(texture_variance, 2)]

# Example usage for a single image:
features = extract_features('dog1.jpg')
print("Extracted [Color, Texture]:", features)
```

```python
image_features = np.array([
# Dogs
[12,18],[14,20],[13,17],[15,19],[11,18],[16,20],
# Cats
[60,65],[62,67],[59,64],[63,66],[61,63],[60,68],
# Cars
[120,20],[118,18],[122,22],[119,19],[121,21],[123,20],
# Mountains
[200,180],[198,182],[201,179],
[203,181],[199,178],[202,183],
# Unknown Images
[250,50], # Flower
[180,20], # Airplane
[170,100], # Bicycle
[90,150], # City
[230,90], # Burger
[180,200] # Beach
])
```

```python
# Create KMeans Model
# ----------------------------------------------------------
kmeans = KMeans(
n_clusters=4,
random_state=42
)
# Train the model
kmeans.fit(image_features)
# Cluster assignments
clusters = kmeans.labels_
```

```python
# Display Results
# ----------------------------------------------------------
print("IMAGE CLUSTERING RESULTS")
for image, cluster in zip(image_names, clusters):
    print(f"{image:18} ---> Cluster {cluster}")
# ----------------------------------------------------------
# Cluster Centers
# ----------------------------------------------------------
print("\nCluster Centers")
print("-"*40)
print(kmeans.cluster_centers_)
```
